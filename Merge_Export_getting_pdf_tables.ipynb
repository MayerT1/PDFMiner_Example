```python
# extract all the tables in the PDF file
all_pdfs = camelot.read_pdf("WWF_I_Appendix_N_3_Table.pdf",pages='0,1,2,3')
# nothing here
all_pdfs[0]
all_pdfs
```

```python
table_0 = all_pdfs[0].df
table_0

table_1 = all_pdfs[1].df
table_1

table_2 = all_pdfs[2].df
table_2

table_3 = all_pdfs[3].df
table_3

table_4 = all_pdfs[4].df
table_4

table_5 = all_pdfs[5].df
table_5
```

```python
merged_tables = pd.concat([table_0, table_1, table_2, table_3, table_4, table_5])
merged_tables
```

```python
# @title 0

from matplotlib import pyplot as plt
import seaborn as sns
merged_tables.groupby(0).size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
merged_tables.to_csv('Appendix_N_3.csv')
```
