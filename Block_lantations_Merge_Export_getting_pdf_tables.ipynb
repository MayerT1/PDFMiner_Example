# Bring in appendix N and I and clean and adjust

```python
app_G = pd.read_csv("Appendix_Ripped_G_Cleaned.csv")
app_G
```

```python
app_N1 = pd.read_csv("Appendix_Ripped_N1_Cleaned.csv")
app_N1
```

```python
Join_tables = app_G.merge(app_N1, on="Location_Site", how = 'inner')
Join_tables
```

```python
Join_tables['Coordinates_N'] = Join_tables['Coordinates_N'].str.replace('\n', '')
Join_tables
```

```python
Join_tables['Coordinates_S'] = Join_tables['Coordinates_S'].str.replace('\n', '')
Join_tables
```

# Fix the header names

```python
Join_tables = Join_tables.drop(['ID_y', 'Region_y', 'Division_y', 'Location/Sub-division_y', 'Zone_y'], axis=1)
Join_tables
```

```python
Join_tables = Join_tables.rename(columns={'Location/Sub-division_x':'Sub_division'})
Join_tables
```

```python
Join_tables = Join_tables.rename(columns={'Ele. \n(m)':'Elev_m', 'No of \nClosures \nverified': 'No_of_Closures_verified ', 'Total \narea \n(ha) Area': 'Total_area_ha', 'area \n(ha) Area \ntraversed \nfor plot \nsampling': 'Area_traversed_for_plot_sampling','Avg. Reg. \nper Plot':'Avg_Reg_Plot','Avg. \nReg. per \nha':'Avg_Reg_per_ha','Total \nRegeneration \nin total area':'Total_Regeneration_total_area'})
Join_tables
```

```python
Join_tables = Join_tables.rename(columns={'Total \narea \n(ha)':'Total_area_ha', 'Area \ntraversed \nfor plot \nsampling': 'Area_traversed_for_plot_sampling'})
Join_tables
```

```python
Join_tables = Join_tables.rename(columns={'Name of VDC':'VDC'})
Join_tables
```

# Switch the Degree Minutes Seconds to Decimal Degrees

```python
Join_tables
```

```python
Join_tables['Coordinates_S_Degree'] = Join_tables.Coordinates_S.apply(lambda x: x[:2])

Join_tables['Coordinates_S_Minute'] = Join_tables.Coordinates_S.apply(lambda x: x[3:5])

Join_tables['Coordinates_S_Second'] = Join_tables.Coordinates_S.apply(lambda x: x[7:])

###
Join_tables['Coordinates_N_Degree'] = Join_tables.Coordinates_N.apply(lambda x: x[:2])

Join_tables['Coordinates_N_Minute'] = Join_tables.Coordinates_N.apply(lambda x: x[3:5])

Join_tables['Coordinates_N_Second'] = Join_tables.Coordinates_N.apply(lambda x: x[7:])
###
Join_tables
```

```python
print("Join_tables before converting")
Join_tables.dtypes
```

```python
Join_tables["Coordinates_S_Degree"] = pd.to_numeric(Join_tables.Coordinates_S_Degree, errors='coerce')
Join_tables["Coordinates_S_Minute"] = pd.to_numeric(Join_tables.Coordinates_S_Minute, errors='coerce')
Join_tables["Coordinates_S_Second"] = pd.to_numeric(Join_tables.Coordinates_S_Second, errors='coerce')


Join_tables["Coordinates_N_Degree"] = pd.to_numeric(Join_tables.Coordinates_N_Degree, errors='coerce')
Join_tables["Coordinates_N_Minute"] = pd.to_numeric(Join_tables.Coordinates_N_Minute, errors='coerce')
Join_tables["Coordinates_N_Second"] = pd.to_numeric(Join_tables.Coordinates_N_Second, errors='coerce')

Join_tables.dtypes
```

# Lambda function to operate on entire columns

```python
Join_tables["Coordinates_S_Decimal"] = Join_tables[["Coordinates_S_Degree","Coordinates_S_Minute","Coordinates_S_Minute"]].apply(lambda row: row.values[0] + row.values[1]/60 + row.values[2]/3600, axis=1)

Join_tables["Coordinates_N_Decimal"] = Join_tables[["Coordinates_N_Degree","Coordinates_N_Minute","Coordinates_N_Minute"]].apply(lambda row: row.values[0] + row.values[1]/60 + row.values[2]/3600, axis=1)


Join_tables
```
